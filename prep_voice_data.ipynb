```python
import os
import sys
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import specgram

import librosa
import librosa.display
from IPython.display import Audio
```

```python
#Read all the audio files into a image and save them
path = './datasets/audio_data/'
out_folder = './datasets/audio_data/out/'

for folder in os.scandir(path):
    if folder.is_dir() and folder.name.startswith('Actor_'):
      for f in os.listdir(path + folder.name + '/'): # go through files in Actor folder
          if(f.endswith(".wav")):
              x, spec = librosa.load(path + folder.name + '/' + f)
              spectrogram = librosa.feature.melspectrogram(y=x, sr=spec, n_mels=128,fmax=8000) 
              db_spec = librosa.power_to_db(spectrogram)
              librosa.display.specshow(db_spec, fmax=8000)
              plt.savefig(out_folder + folder.name + '_'+ f + '.jpeg')
```
